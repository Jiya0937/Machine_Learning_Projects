# Sets


```python
marks={56,48,37,93,39,86,22,44}
print (marks)
```

```python
# not print duplicats value
student={"aditya","sanadhya","jiya","jiya"}
print(student)
```

```python
#add function(add single value)
student={"aditya","sanadhya","jiya","jiya"}
student.add("riya")
print(student)
```

```python
#undate function(add multiple values)
nums={1,2,3,4}
nums.update((9,8,7))
print (nums)
```

```python
#Remove function (remove only 1 element)
nums={1,2,3,4,9,7,8}
nums.remove(9)
print (nums)
```

```python
#pop (remove random element)
nums={1,2,3,4,9,7,8}
nums.pop()
print (nums)
```

```python
#clear function
nums={1,2,3,4,9,7,8}
nums.clear()
print (nums)
```

```python
#union function(combine 2 sets)
nums1={1,2,3,4}
nums2={10,9,7,8}
print(nums1|nums2)
```

```python
#intersection function(print common element)
nums1={1,2,3,4,5,6}
nums2={4,5,6,7}
print(nums1&nums2)
```

```python
#difference function(print only element that is not present in num2 and print only nums1 element )
nums1={1,2,3,4,5,6}
nums2={4,5,6,7}
print(nums1-nums2)
```

```python
#symmentric distance(remove common element)
nums1={1,2,3,4,5,6}
nums2={4,5,6,7}
print(nums1^nums2)
```

```python
#membership test
nums1={1,2,3,4,5,6}
print(5 in nums1)
print(9 in nums1)
```

# Dictionary

```python
#print dictionary
intro={
    "name"   :"jiya",
    "age"    :18,
    "collage":"acropolis"
    
}
print(intro)
```

```python
#print specific value in any key
intro={
    "name"   :"jiya",
    "age"    :18,
    "collage":"acropolis"
    
}
print(intro["age"])
```

```python
#add keys and values in empty dictionary
subject={}
subject={"physics":89,'chemistry':94,"maths":90}
print(subject)
```

```python
#Adding one more key and value

subject={"physics":89,'chemistry':94,"maths":90}
subject["english"]=78
print(subject)
```

```python
#edit or update element

subject={"physics":89,'chemistry':94,"maths":90}
subject["physics"]=90
subject["chemistry"]=100
print(subject)
```

# removing items in dictionary

```python
# pop(key)
subject={"physics":89,'chemistry':94,"maths":90}
subject.pop("physics")
print(subject)
```

```python
# delete item
subject={"physics":89,'chemistry':94,"maths":90}
del subject["physics"]
print(subject)
```

```python
# popitem() delete last item
subject={"physics":89,'chemistry':94,"maths":90,"english":101}
subject.popitem()
print(subject)
```

# looping through a dictionary

```python
student={"riya :":"not adult","jiya :":" adult","sandhya :":"adult","aditya :": "aditya"}
for riya in student:
    print(riya,student[riya])
```
