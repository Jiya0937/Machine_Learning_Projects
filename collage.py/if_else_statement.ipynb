```python
unit=int(input("enter unit : "))
if unit<=100:
    bill=unit*5
elif unit<=200:
    bill=100*5+(unit-100)*7
elif unit<=300:
    bill=100*5+100*7+(unit-200)*10
else:
    bill=100*5+100*7+100*10+(unit-300)*15
print("Total Bill : ",bill)
```
