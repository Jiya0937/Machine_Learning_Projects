```python
try:
    a = int(input("value of a : "))
    b = int(input("value of b : "))
  
    print(a/b)
except (ArithmeticError,ZeroDivisionError,ValueError):
    print("This value is not valid")
print("Thank you😊")
```

```python
# finally : it is finally block that excecute always if error created or noot in any condition
try:
    a = int(input("value of a : "))
    b = int(input("value of b : "))
  
    print(a/b)
except ZeroDivisionError:
    print("invalid input")
finally:
    print("database connection close ")
    
```

```python
import os
try: 
    print("try")
    os._exit(0)
except ValueError:
    print('except')
finally:
    print('finally')
```

```python

#nested try
try:
    print("outer try")
    try:
        print("inner try")
        a = int(input())
        b = int(input())
        print(a/b)
    except ZeroDivisionError:
        print("value is not valid")
    except ValueError:
        print("value is not valid")
    finally : 
        print("inner try database is close")
except ZeroDivisionError:
    print("value is not valid")
finally :
    print("outer tey database is close")
```

```python
#else block : if any error is not created so else bllock run otherwise not run
try: 
    print("try")
    a = int(input())
    b = int(input())
    print(a/b)
except ZeroDivisionError:
    print("Zero Division error not valid value")
else: 
    print("else block run") # error is not create so else run
finally: 
    print("finally block")
```
