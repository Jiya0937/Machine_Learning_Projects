3. Find 2nd highest number and 2nd minimum from List of n elements.

4. Separate even and odd numbers from a list.

5. A list is considered a palindrome if it reads the same forward and backward. Write a program to check if a given list is a palindrome.

6. Given a list of consecutive numbers with one missing number, find the missing number.

```python
#1 solution
numbers = [1, 2, 3, 2, 4, 5, 6, 4, 7, 8, 8, 9]

unique_elements = set()
duplicates = set()

for item in numbers:
    if item in unique_elements:
        duplicates.add(item)
    else:
        unique_elements.add(item)

print("Total duplicate elements:", len(duplicates))
```

```python
#2 solution
numbers = [1, 2, 3, 2, 4, 5, 3, 6]
unique_numbers = list(set(numbers))
print("List after removing duplicates:", unique_numbers)
```

```python
#3 solution
numbers = [10, 5, 8, 20, 15, 3, 25, 12]

# Remove duplicates and sort the list
unique_numbers = sorted(set(numbers))

# Check if the list has at least 2 elements
if len(unique_numbers) < 2:
    print("Not enough unique numbers in the list")
else:
    second_min = unique_numbers[1]  # Second minimum is at index 1
    second_max = unique_numbers[-2]  # Second highest is at index -2

    print("Second Minimum:", second_min)
    print("Second Highest:", second_max)
```

**Function Excercise**

1. Write a Python function is_anagram(str1, str2) that takes in two strings and returns True if the strings are anagrams of each other, and False otherwise. An anagram is a word or phrase formed by rearranging the letters of a different word or phrase, typically using all the original letters exactly once.

2. Write a function count_vowels(s) that counts how many vowels (a, e, i, o, u) appear in a string.
Ignore case (so A and a are both vowels).

3. Write a function that finds and returns the longest word in a sentence.

4. Given numbers 1 to n with one missing, find the missing number using XOR.

5. Write a function that returns the first character that doesn’t repeat in a string.
Example:
"stress" → 't'
"aabbcc" → None

```python
#Excercise 1  O(nlogn)
def is_anagram(str1, str2):
    # First, check if the lengths of the strings are the same
    if len(str1) != len(str2):
        return False

    # Sort both strings and compare if they're the same
    return sorted(str1) == sorted(str2)

# Test cases
print(is_anagram("listen", "silent"))  # Expected output: True
print(is_anagram("hello", "world"))    # Expected output: False
print(is_anagram("evil", "vile"))      # Expected output: True
```

```python
#Excercise 1 O(n)
from collections import Counter

def is_anagram(str1, str2):
    # Check if the frequency counts of characters in both strings are the same
    return Counter(str1) == Counter(str2)

# Test cases
print(is_anagram("listen", "silent"))  # Expected output: True
print(is_anagram("hello", "world"))    # Expected output: False
print(is_anagram("evil", "vile"))      # Expected output: True
```

```python
#Excercise 2
def count_vowels(s):
    vowels = "aeiou"
    count = 0
    for char in s.lower():
        if char in vowels:
            count += 1
    return count

# Test cases
print(count_vowels("Python Programming"))  # 4
print(count_vowels("HELLO"))               # 2
```

The key parameter lets you decide what “largest” means by applying a function to each element before comparison.

Here:

words → is a list of strings, e.g. ["Python", "is", "amazing"]

key=len → tells Python to compare the lengths of the strings instead of the strings themselves.

So, this expression:

max(words, key=len)

means:
"From the list words, find the word with the maximum length."

```python
#Excercise 4
def find_missing_xor(lst):
    n = len(lst) + 1
    xor_all = 0
    xor_lst = 0
    for i in range(1, n + 1):
        xor_all ^= i
    for num in lst:
        xor_lst ^= num
    return xor_all ^ xor_lst
print(find_missing_xor([1, 2, 4, 5, 6]))  # 3
```

**Decorator Function**

A decorator function in Python is a higher-order function that modifies the behavior of another function or method. Decorators are widely used to add extra functionality to existing functions without altering their structure

```python
#Simple Example 1
def my_decorator(func):
    def wrapper():
        print("Something is happening before the function is called.")
        func()
        print("Something is happening after the function is called.")
    return wrapper

@my_decorator
def say_hello():
    print("Hello!")

say_hello()
```

```python
#Simple Example 2
def smartdivision(func):
  def inner(a,b):
    if b==0:
      print('division not possible')
    else:
      return func(a,b)
  return inner

@smartdivision
def division(a,b):
  return a/b
print(division(10,2))
print(division(10,0))
```

```python
#Calling Parent Method Using super()
class Animal:
    def __init__(self, species):
        self.species = species

    def make_sound(self):
        print("Some generic animal sound")

class Dog(Animal):
    def __init__(self, name, species="Dog"):
        super().__init__(species)
        self.name = name

    def make_sound(self):
        super().make_sound()  # Call parent method
        print(f"{self.name} barks!")

dog = Dog("Buddy")
dog.make_sound()
```

```python
#Polymorphism with Inheritance
class Shape:
    def area(self):
        print("Area calculation not defined.")

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return 3.14 * self.radius ** 2

class Rectangle(Shape):
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height

shapes = [Circle(5), Rectangle(4, 6)]
for shape in shapes:
    print(f"Area: {shape.area()}")
```

After completing write operation highly recommended to close the file. But if i dont want to close, it should be closed automatically, and i want to group all the file i/o related operation as a block and the file should be closed automatically once the operation complete. then we should go for 'with' statement

  with open('filename','mode') as f:
