```python
s={1,2,3,4,5,6}

s1=s  # (shallow copy) reference and address of s and s1 is same(because in python same type of data store in only one variable and one location)
s2=s.copy()  #  copy ad address of s and s2 is difference

print(s1 is s)
print(s2 is s)
```

```python
# union
nums1={1,2,3,4}
nums2={10,9,7,8}
print(nums1.union(nums2))
print(nums1|nums2)
```

```python
#difference 
nums1={1,2,3,2,11}
nums2={2,4,6,8,1}
print(nums1.difference(nums2))
print(nums1-nums2)
```

```python
# set comprehension
s={1,2,3,4,4,5,6}
s={x*x for x in range(1,6)}
print(s)
```
