```python
# 2. Write a program to calculate and display the sum of even and sum of odd numbers in a list of integers.

l = eval(input("enter list"))
print(l)
e=[x for x  in l if x%2==0]
o=[x for x  in l if x%2!=0]
print("even number : ",e)
print("odd number  : ",o)
```
