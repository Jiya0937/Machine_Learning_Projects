```python
class Teacher:
    def _init_ (self,name,dep,sub):
        name = self.name
        dep = self.dep
        sub = self.sub
    def show(self):
        print(f"the name teacher is {self.name}").title()
        print(f"department of the teacher is {self.dep}").title()
        print(f"subject of the teacher is {self.sub}").title()
t1 = Teacher("Jiya","AIML","Python")
t1.show()
        
```

```python
#single level inheritance

class parent:
    def display(self):
        print("This is Parent Class")
class child(parent):
    def show(self):
        print("I can access Parent Class")

c1 = child()
c1.display()
c1.show()
```

```python
#2. Multiple Inheritance
class parent:
    def display(self):
        print("This is first Parent Class")
class parent2:
    def show(self):
        print("This is second parent class")
class child(parent, parent2):
    def show2(parent):
        print("I can access both parent1 and parent2")

c1 = child()
c1.display()
c1.show()
c1.show2()
```

```python
#. multilevel inheritance
class Grandparent:
    def display(self):
        print("This is a Grand Parent Class")
class parent(Grandparent):
    def show(self):
        print("I can access grand Parent Class")
class child(parent):
    def show2(parent):
        print("I can access both parent class and grand parent class")
        
s1 = child()
s1.display()
s1.show()
s1.show2()
```

```python
#hierarchical inheritance : one parent class and multiple child class

class parent:
    def display(self):
        print("I am a parent class")
        
class child1(parent):
    def show1(self):
        print("This is a first child Class")
        
class child2(parent):
    def show2(self):
        print("This is a second child class")
        
class child3(parent):
    def show3(self):
        print("This is third child class")
#child one        
c1 = child1()
c1.display()
c1.show1()
```

```python
#hybrid Inheritance

class base:
    def display(self):
        print("this is base class")
class parent1(base):
    def show1(self):
        print("this is parent 1")
class parent2(base):
    def show2(self):
        print("this is parent 2")
class child(parent1, parent2):
    def child_method(self):
        print("this is child class")

c1 = child()
c1.display()
c1.show1()
c1.show2()
c1.child_method()
```
