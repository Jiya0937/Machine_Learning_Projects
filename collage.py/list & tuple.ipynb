```python
# check frequency
ele = [1,3,5,3,1,3]
count=1
freq={}
for i in ele:
    if i in freq:   
        freq[i]=count
        count+=1
    else:
        freq[i]=count
print(freq)
```

```python
# check duplicates
ele = [1,3,5,3,1,3]
unique={}
duplicate={}
for i in ele:
    if i in unique:
        duplicate.add(i)
    else:
        unique.add(i)
```
