```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("C:\\Users\\jiyad\\OneDrive\\Documents\\Desktop\\ML_task1_dataset.csv")
df
```

```python
x = df[['Hours']]  #double brackets  = 20 input
y = df['Scores']   # target colunm

model = LinearRegression()
model.fit(x, y)
predicted_score = model.predict(x)

#evaluate
mae = mean_absolute_error(x,predicted_score)
mse = mean_squared_error(x,predicted_score)
rmse = np.sqrt(mse)

#show result
print("mean absolute error : ",mae)
print("mean squared error : ",mse)
print("root mean squared error : ",rmse)

new_hours = float(input("enter your studies hours : "))
new_pred = model.predict([[new_hours]])

print("you will get marks according to your study hours likely : ",new_pred[0])
```

```python
student = {
    'hours' : [1,2,3,4,5,6,7],
    'week' : [5,10,20,25,48,45,50],#double brackets  = 20 input
    'sleep time':[9,8,7,7,6,6,4],
    'score' : [32,45,54,67,77,89,95]   # target colunm
}
df = pd.DataFrame(student)
df
x = df[['hours','week','sleep time']]
y = df['score']
model = LinearRegression()
model.fit(x,y)
predicted_score = model.predict(x)

#evaluate
mae = mean_absolute_error(y,predicted_score)
mse = mean_squared_error(y,predicted_score)
rmse = np.sqrt(mse)

#show result
print("mean absolute error : ",mae)
print("mean squared error : ",mse)
print("root mean squared error : ",rmse)

new_hours = float(input("enter your studies hours : "))
new_week_hours = float(input("enter your studies hours in 1 week : "))
new_sleep_hours = float(input("enter your sleep hours : "))

new_pred = model.predict([[new_hours,new_week_hours,new_sleep_hours]])

print("you will get marks according to your study hours likely : ",new_pred[0])
```
