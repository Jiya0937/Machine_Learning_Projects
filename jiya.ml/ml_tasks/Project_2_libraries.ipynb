# Now start to find outliers

```python
# calculate mean value of all numerical columns
mean={}
skip_col=["dteday","season","hr","holiday","mnth","yr"]
print("======mean of all numeric columns=======")
for i in all_combine.columns:
    if  i in skip_col:
        pass
    else:
        if pd.api.types.is_numeric_dtype(all_combine[i]):
            mean[i]=all_combine[i].mean()
print(mean)
```

```python
#calculate varience
std={}
skip_col=["dteday","season","hr","holiday","mnth","yr"]
print("======varience of all numeric columns=======")
for i in all_combine.columns:
    if  i in skip_col:
        pass
    else:
        if pd.api.types.is_numeric_dtype(all_combine[i]):
            std[i]=all_combine[i].std()
print(std)
```

```python
neg={}
skip_col=["dteday","season","hr","holiday","mnth","yr"]
for i in all_combine.columns:
    if i in skip_col:
        pass
    else:
        neg[i]=mean[i]-3*std[i]
print(neg)                  
```

```python
pos={}
skip_col=["dteday","season","hr","holiday","mnth","yr"]
for i in all_combine.columns:
    if i in skip_col:
        pass
    else:
        pos[i]=mean[i]+3*std[i]
print(pos)  
```

```python
skip_col=["dteday","season","hr","holiday","mnth","yr"]
for i in all_combine.columns:
    if i in skip_col:
        pass
    else:
        outlier=all_combine[(all_combine[i]>pos[i]) | (all_combine[i]<neg[i])]
outlier        
```

```python
outlier.shape
```

```python
# find no outlier values
skip_col=["dteday","season","hr","holiday","mnth","yr"]
for i in all_combine.columns:
    if i in skip_col:
        pass
    else:
        no_outlier=all_combine[(all_combine[i]<pos[i]) | (all_combine[i]>neg[i])]
no_outlier    
```
