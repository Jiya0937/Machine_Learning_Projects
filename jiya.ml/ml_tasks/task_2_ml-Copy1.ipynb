# Predict Scores of Student Accouding to Their Academic Perfomance and Sleep Hours

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

```

```python
data = pd.read_csv("")
data
```

```python
data.isnull().sum()
```

```python
df = data.copy()
df
```

```python
from sklearn.model_selection import train_test_split

X = df[['hours_studied', 'sleep_hours', 'attendance_percent', 'previous_scores']]
y = df[['exam_score']]

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.2 , random_state=2)

print("===Traning Data===\nschool perfomance")
print(X_train)

print("\n----Testing Data----\nschool perfomance")
print(X_test)

print("\n===Traning Data===\nExam Scores")
print(y_train)

print("\n----Testing Data----\nExam Scores")
print(y_test)
```

```python
# Predict the Final Scores According to Academic Perfomance of Student 
X = X_train[['hours_studied', 'sleep_hours', 'attendance_percent' ,'previous_scores']]
y = y_train['exam_score']

model = LinearRegression()
model.fit(X,y)
prediction = model.predict(X)

# Find Errors between Actual Scores and Predicted Scores
mae = mean_absolute_error(y,prediction)
mse = mean_squared_error(y,prediction)
rmse = np.sqrt(mse)
r2 = r2_score(y,prediction)

print("Mean Absolute Error : " , round(mae,2))  # round remove large decimal value after point
print("Mean Squared Error : " , round(mse, 2))
print("Root Mean Squared Error : " , round(rmse, 2))
print("R^2 Score (Model Accuracy) : ", round(r2, 2)) # if it r^2  closer to 1 = Better

# histogram
plt.figure(figsize = (10, 6))  # size of the graph(height , length)
plt.hist(y_train["exam_score"], bins=30, color='skyblue', edgecolor='black')
plt.title("Distributionof Final Exam Scores")
plt.xlabel("Final Exam Score")
plt.ylabel("Number of Student")
plt.grid(True)  # background lines
plt.show()

# scatter + Regression line
plt.figure(figsize = (10,6))
plt.scatter(X["hours_studied"], y, color = 'blue', label = "Actual Scores")
plt.plot(X["hours_studied"],prediction, color = 'red', label = "Predicted Scores (Regression Line)")
plt.title("Model Prediction VS Actual Score")
plt.xlabel("Study Hours per Week")
plt.ylabel("Final Output")
plt.grid(True)
plt.show()

new_perfomance = X_test[['hours_studied', 'sleep_hours', 'attendance_percent' ,'previous_scores']]
new_score_prediction = model.predict(new_perfomance)


print("\n--------Prediction Output of Student For Exam Scores -----------")
df2 = pd.DataFrame(new_score_prediction)
df2



```
