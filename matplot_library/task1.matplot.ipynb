```python
#task2 
"""�
� Problem Description: 
A company wants to compare its market share percentage with competitors. Interviewers 
might test if you know pie charts and percentages. """

companies=['A','B','C','D']
shares=[35,25,20,20]

plt.pie(shares,labels=companies,autopct="%1.1f%%",)
startangle=0
plt.title("Market current condition")
plt.show()
```

```python
"""3. Product Category Sales 
�
� Problem Description: 
A retail manager wants to compare sales across product categories. Interviewers might ask 
which chart type best compares groups."""
catagorie=["Electronics","Furniture","Grocery","Clothing"]
sales=[5000,3000,7000,4000]
plt.bar(catagorie,sales)
plt.title("current situation of sales in market".title())
```

```python
"""5. Stock Price Movement 
�
� Problem Description: 
Financial analysts want to see how two companies’ stock prices fluctuate during the week. 
Interviewers might ask how to plot multiple lines on one graph."""
days=["Mon","Tue","web","Thu","Fri"]
company_A=[120,125,123,128,130]
company_B=[90,92,91,95,94]

plt.plot(days,company_A,label="Comapany A",marker="o")#draw graph
plt.plot(days,company_B,label="Company B",marker="s")#draw graph
plt.title("stock price movement".title())#title on holl graph
plt.xlabel("Days")#title in x axis
plt.ylabel("Price($)")#title in y axis
plt.legend()#legent use to label showing
plt.grid()
plt.show()

```

```python
"""6. Employee Performance Analysis 
� Problem Description: 
An HR team wants to evaluate employees across different skills. Interviewers often ask 
about radar charts (spider plots). """
skills=["comminication","tech","leadership","Teamwork","creativity"]
scores=[7,9,8,9,3]
plt.radar(skills)
```

```python
'''7. Customer Age Distribution 
�
� Problem Description: 
A business wants to understand customer age demographics. Interviewers often ask when 
to use histograms. '''
Ages= [22,25,29,21,34,42,37,31,28,40,45,36,30,32,25,27,29,33,41]
#bins means how many parts in which  ggraph divided
# color: show which color show in graph
#egde : show which color egde show in graph  
plt.hist(Ages,bins=5,color="skyblue",edgecolor='black')
plt.title("Customer Age Distribution")
plt.xlabel("Age Range")
plt.ylabel('Count')
plt.show()
```

```python
"""8. Sales vs Profit Analysis 
� Problem Description: 
A business analyst wants to study the relationship between sales and profit. Interviewers 
may check knowledge of scatter plots for correlation."""
sales=[1000,2000,3000,4000,5000,6000,7000]
profit=[100,400,800,900,1200,1500,2000]
plt.scatter(sales,profit,color="red",marker="o")
plt.title("sales vs profit analysis")

plt.grid()
plt.xlabel("Sales")
plt.ylabel("Profit")
plt.show()
```
