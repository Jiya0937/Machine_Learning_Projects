# Import libraries

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

# Data Collectionn and analysis

```python
df = pd.read_csv("diabetes.csv")
df.sample(5)
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
df['Outcome'].value_counts()

# 0 -> Non Diebetes
# 1 -> Diebetes
```

# train test split

```python

X = df.drop(columns = 'Outcome', axis = 1)
Y = df['Outcome']
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.3, random_state = 2)
    
```

```python
X_train.head()
```

```python
print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

# Standerization 

```python
scaler = StandardScaler()

scaler.fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
X_train_scaled
```

```python
X_train_data = pd.DataFrame(X_train_scaled, columns = X_train.columns)
X_test_data = pd.DataFrame(X_test_scaled, columns = X_test.columns)
```

```python
X_train_data.head()
```

# Train the Model

```python
classifier = svm.SVC(kernel = 'linear')

# training the support vector Machine Classifier
classifier.fit(X_train, y_train)

```

# Model Evaluation

```python
# accuracy of the trianed data
X_train_prediction = classifier.predict(X_train)

# find Accuracy of the trianing data
training_Data_accuracy = accuracy_score(X_train_prediction, y_train)
training_Data_accuracy
```

```python
# accuracy of the testing data
X_test_prediction = classifier.predict(X_test)

# find the Acuuracy Score of the testing data
testing_data_accuracy = accuracy_score(X_test_prediction, y_test)
testing_data_accuracy
```

```python
# Making the input data
input_data = 1,85,66,29,0,26.6,0.351,31

# changing the input_Data to numpy array
input_data_as_numpy_array = np.asarray(input_data)    

input_data_reshaped = input_data_as_numpy_array.reshape(1, -1)
std_data = scaler.transform(input_data_reshaped)


# Input data prediction
input_Data_prediction = classifier.predict(std_data)
if input_Data_prediction == 1:
    print("yes, patient has a diebetes")
else:
    print("No, patient does not have diebetes")
```
