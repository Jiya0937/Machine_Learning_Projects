```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OneHotEncoder
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
```

```python
house_data = pd.read_csv("data.csv")
house_data.sample(5)
```

# Features Analysis

```python
house_data.info
```

```python
house_data.describe()
```

```python
house_data.isnull().sum()
```

```python
print(house_data["waterfront"].value_counts())
print(house_data["country"].value_counts())
print(house_data["city"].value_counts())
```

# Drop unneccesary Columns

```python
house_data = house_data.drop(columns = ["date", "country", "street"])
```

# Handle StateZip columns which has Alphanumerical Value

```python
house_data[['state', 'zip']] = house_data['statezip'].str.split(' ',expand = True)
house_data = house_data.drop(columns = ['statezip', 'state'])
house_data
```

# Standarization

```python
scaler = StandardScaler()
numerical_column = ["bedrooms",	'bathrooms', 'sqft_living',	'sqft_lot',	'floors', 'waterfront',	'view',	'condition', 'sqft_above',	'sqft_basement',	'yr_built',	'yr_renovated', 'zip'	]

house_data[numerical_column] = scaler.fit_transform(house_data[numerical_column])
house_data
```

# Get Dummies of the Different cities

```python
counts = house_data['city'].value_counts()
```

```python
house_data['city'].nunique()
threshold = 50
```

```python
replace_remaining = counts[counts <= threshold].index
```

```python
city_encoded = pd.get_dummies(house_data['city'].replace(replace_remaining, "Other"),dtype = int)
```

```python
city_encoded
```

```python
print(house_data.shape)
print(city_encoded.shape)
```

```python
# Merse Main Dataset of house that is house_data and city encoded data that is city_encoded
```

```python
house_data = house_data.reset_index(drop = True)
city_encoded = city_encoded.reset_index(drop = True)
```

```python
house_data = pd.concat([house_data, city_encoded], axis = 1)
```

```python
house_data = house_data.drop(columns = ['city'])
```

# Train Test split


```python
X_train, X_test, y_train, y_test = train_test_split((house_data.drop(columns = ["price"])),house_data['price'], test_size = 0.3, random_state = 45)
```

```python
X_train.head(5)
```

```python
# Model Traning 
```

```python
model = LinearRegression()

model.fit(X_train, y_train)
```

```python
# Model Evalution
```

```python
X_train_prediction = model.predict(X_train)
mae_train = mean_absolute_error(y_train, X_train_prediction)
mae_train
```

```python
X_test_prediction = model.predict(X_test)
mae_test = mean_absolute_error(y_test, X_test_prediction)
mae_test
```

```python
from sklearn.metrics import r2_score

r2_train = r2_score(y_train, X_train_prediction)
r2_test = r2_score(y_test, X_test_prediction)

print(r2_train)
print(r2_test)
```
