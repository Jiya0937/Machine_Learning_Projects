```python
#positional 
def show(name,age,gf):
    print('helloe',name,'your age is',age,'your gf is')
show('manav',10,'pooja')
    
```

```python
#default (condition-1)
def show(name,age,gf="khushi"):
    print('helloe',name,'your age is',age,'your gf is')
show('harsh',21,'mahak')
```

```python
#default (condition-2) when defaul more than 2 argument is not exepted 
#conclusion - dafault argument should be mention on the last
def show(name,age=17,gf="khushi"):
    print('helloe',name,'your age is',age,'your gf is',gf)
show('harsh','ram')
```

```python
#variable length argunment(in which length of argument is not fixed)
#syntax = def sum(*n):   # in which * denoted not fix value and n denoted tuple
#example
def sum(*n):
    sum=0
    for i in n:
        sum=sum+i
    print('sum :',sum)

sum(12, 3, 4)
```

```python
#first argument should be positional argument and other should be variable length argunment
def marks(name,*marks):
    sum=0
    for i in marks:
        sum=sum+i
    print('hello',name,'your Total marks :',sum)
marks("jiya",89,90,90)
marks('riya',90,98,97)

    
```

```python
def marks(*marks,name):
    sum=0
    for i in marks:
        sum=sum+i
    print('hello',name,'your Total marks :',sum)
marks(89,90,90,name="jiya")   # here keyword argument that is name=jiya
marks(90,98,97,name='riya')
```

# variables


```python
# local variable
'''A local variable is a variable that is declared inside a
function or block and can only be used within that function.
It exists only while the function is running — 
once the function ends, the variable is destroyed.'''
```

```python
# Global variable
'''A global variable is declared outside all functions, usually at the top of the program.
It can be accessed and modified by any function in the program.'''
```

```python
# global and local variable 
a=10 # global variable
def f1():
    global a
    a=777  # local variable
    print("number : ",a)
def f2():
    print("numder : ",a)
f2()
f1()
```

```python
# Anonymous Function / Lambda Function
'''An anonymous function is a function without a name.
In most languages (like Python, JavaScript, etc.), anonymous functions are used for short, 
one-time tasks where creating a full def or function block is unnecessary.'''
# syntax := lambda input:expression

#normal function
def square(n):
    return n*n
square(2)

```

```python
#lamda function
s=lambda n:n*n
print(s(2))
```

```python
#find modulas
f=lambda a,b:a/b
f(10,2)
```

```python
#WAP to find sum of 2 numbers
s=lambda a,b:a+b
s(2,3)
```
