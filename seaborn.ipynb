```python
#bbox_to_anchor=(1.05,1)  -> pushes legent slightly outside outide to the right
#bbox_to_anchor=(05,-0.1)  -> pushes legent below the plot , centereds
#bbox_to_anchor=(0,1)  with loc="upper ight"puts legend at top-left outside

sns.lineplot(x="day",y="total_bill",data=df,hue="sex",style="time",palette="RdBu")
plt.legend(loc="upper left",bbox_to_anchor=(1.05,1))
```

```python
# Bar Chart 
# A bar chart is a graph that uses rectangular bars to show the values of difference
# We use bar charts to compare categories easily and see which one has the height
b=sns.load_dataset("titanic")
b.head()
```

```python
sns.barplot(x="who",y="fare",data=b)
plt.show()
```

```python
sns.barplot(x="who",y="fare",data=b,color="magenta")
plt.show()
```

```python
# ci hide or remove the line that show between the bar  chart
sns.barplot(x="fare",y="who",data=b,ci=False)
plt.show()
```

```python
# estimator set mean mode and mediun
from numpy import mean
sns.barplot(x="who",y="fare",data=b,estimator=mean)
plt.show()
            
```

```python
sns.barplot(x="who",y="fare",data=b,facecolor=[1,1,1,0],edgecolor="0.1")
plt.show()
```

```python
# HeatMap
# A heatMap is a chart that shows data values in a matrix form , where colors represent
```

```python
import numpy as np
data = np.random.randint(low=1,high=100,size=(10,10))
# random.randint give the random value according that parameter we will passed  
data
```

```python
sns.heatmap(data=data)
```

```python
# cmap help to show combination color
sns.heatmap(data=data,cmap="GnBu",alpha=0.5)
```

```python
# linecolor give outline of the graph
# linewidth give thickness of the border
sns.heatmap(data=data,linecolor="yellow",linewidth="2")
```

```python
# annot= True help to show the actual value of the graph
sns.heatmap(data=data,annot=True)
```

```python
# cbar = False is remove the actuala value of the graph
sns.heatmap(data=data,cbar=False)
```

```python
# x ticklabels and y ticklabels remove the x and y labels
sns.heatmap(data=data,xticklabels=False,yticklabels=False)
```

```python
# Violin plot
# violin plot is a graph that shows the sidtribution data for different 
# It combines a box plot ( showing mediun & quartiles ) with a KDE curve  ( showing the
# KDE stands for kernel density estimation
# a KDE plot is a smoothed curve that shows the probability distribution of a dates 
```

```python
df=sns.load_dataset("tips")
df.head()
```

```python
"""A violin plot in Seaborn is a type of data visualization that shows the distribution of a numeric variable for one or more categories.
It’s similar to a box plot, but instead of just showing summary statistics (like median, quartiles, and outliers), it uses kernel density estimation
(KDE) to display the probability density of the data at different values. This gives it the "violin" shape."""

sns.violinplot(x=df["total_bill"])
```

```python
# Draw one violin for each category in sex.
sns.violinplot(x="sex",y="tip",data=df)
```

```python
sns.violinplot(x="sex",y="tip",data=df,hue="day")
```

```python
sns.violinplot(x="day",y="tip",data=df,hue="sex",split=True)
```
