```python
# ------------------------------------------------------------
# 📦 STEP 1: Import Required Libraries
# ------------------------------------------------------------
import pandas as pd
import os

# ------------------------------------------------------------
# ⚙️ STEP 2: Load the Dataset
# ------------------------------------------------------------
# Change this path to where your CSV file is located
df = pd.read_csv("C:\\Users\\jiyad\\Downloads\\padchat.ai.zip")

print("Original dataset shape:", df.shape)
print(df.head())

# ------------------------------------------------------------
# 🧹 STEP 3: Basic Cleaning
# ------------------------------------------------------------

# 1️⃣ Keep only useful columns
df = df[['ImageID', 'Labels', 'Projection', 'PatientAge', 'PatientSex']]

# 2️⃣ Remove rows with missing or empty labels
df = df.dropna(subset=['Labels'])

# 3️⃣ Keep only standard 'PA' (Posterior-Anterior) X-rays
df = df[df['Projection'] == 'PA']

# 4️⃣ Create columns for diseases you want to detect
df['Pneumonia'] = df['Labels'].str.contains('pneumonia', case=False, na=False).astype(int)
df['TB'] = df['Labels'].str.contains('tuberculosis', case=False, na=False).astype(int)

# 5️⃣ Keep only rows where either Pneumonia or TB is present
df = df[(df['Pneumonia'] == 1) | (df['TB'] == 1)]

# 6️⃣ Remove duplicate image entries
df = df.drop_duplicates(subset=['ImageID'])

print("After cleaning:", df.shape)
print(df.head())

# ------------------------------------------------------------
# 📁 STEP 4: Verify image files exist (optional)
# ------------------------------------------------------------
# If you have a folder with all images, you can check if each file exists
img_dir = "/content/PadChest/images"  # change path accordingly

df['ImagePath'] = df['ImageID'].apply(lambda x: os.path.join(img_dir, f"{x}.jpg"))
df = df[df['ImagePath'].apply(os.path.exists)]

print("Valid images found:", len(df))

# ------------------------------------------------------------
# 💾 STEP 5: Save Cleaned Data
# ------------------------------------------------------------
df.to_csv("/content/PadChest_cleaned.csv", index=False)
print("✅ Cleaned dataset saved as PadChest_cleaned.csv")
```
