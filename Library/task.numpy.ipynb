```python
import numpy as np
```

```python
## 1. Student Marks Analysis (Education Domain)

maths=np.array([84,86,74])
science=np.array([67,82,98])
english=np.array([77,89,43])
total=[]
print("=========each student’s average==========\n")
for i in range(0,3):
#task 1
    s=maths[i]+science[i]+english[i]
    arr=np.array([maths[i],science[i],english[i]])
    print(f"student{i+1}  : marks{arr}")
    ave=np.mean(arr)
    print("1. presantege :",ave)
    total.append(ave)
    
#task 2
print("\n========average of per subject========")
print("average of maths : ",np.mean(maths))
print("average of science : ",np.mean(science))
print("average of english : ",np.mean(english))

#task 3 
print("\n======topper stusent======") 
print("topper student marks : ",np.max(total))

#task4
print("")
for i in range(0,len(total)):
    
    if 100>total[i]>=90:
        print(f"student {i+1} grade:{'A'}")
    elif 90>total[i]>=75:
        print(f"student {i+1} grade:{'B'}")
    elif 75>total[i]>=33:
        print(f"student {i+1} grade:{'C'}")
    elif 33>total[i]>=0:
        print(f"student {i+1} grade:{'D'}")  
```
