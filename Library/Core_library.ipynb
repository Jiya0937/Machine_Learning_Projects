# STATISTICS LIBRARY

```python
"""The statistics library provides functions for calculating mathematical statistics of numeric data.
It is useful for data analysis and statistics-related tasks."""
```

```python
import statistics as st
```

```python
#mean
DATA=[10,20,30,40,50,60,70]
st.mean(DATA)
```

```python
#mediun
st.median(DATA)
```

```python
#mode
DATA=[10,20,30,40,50,10,50,60,70,50]
st.mode(DATA)
```

```python
#STANDERD  DEVIATION
st.stdev(DATA)
```

```python
#varience
st.variance(DATA)
```

```python
import datetime as dt
```

```python
#current date
now=dt.datetime.now()
```

```python
print(now)
```

```python
#date
date=dt.date(2025,8,25)
date
```

```python
#time
time=dt.time(5,23,9)
time
```

# timedelta

```python
#date with time
now=dt.datetime.now()
print(now)
```

```python
# Add 5 days or multiple months and year
future = now + dt.timedelta(days=6)
print("After 6 days : ",future)
```

```python
#substract 2 hour
past=now - dt.timedelta(hours=2)
print("2 hours left : ",past)
```

# string formate time

```python
#formating
""" y=year=25
Y=year=2025
d=day
m=month
I=hour
M=minute
p=am or pm
A=day"""
```

```python
# 2025-08-25
print(now.strftime("%y-%m-%d"))
```

```python
#19/08/2025
print(now.strftime("%d/%m/%y"))
```

```python
#August 19,2025
print(now.strftime("%B %d, %Y"))
```

```python
#05:35 PM
print(now.strftime("%I:%M %p"))
```

```python
print(now.strftime("%A"))
```

# Collection library

```python
import collections as c
```

```python
# count fequency of element in a collection
data=[2,3,4,5,6,67]
print(c.Counter(data))
```

```python
# creates tuple like objects with named fields
point = c.namedtuple("point",['x','y'])
p=point(10,20)
print(p.x,p.y)
```

```python
j=c.deque(data)
j.appendright(70)
print(j)
```

```python
# Double ended queue , allow fast appends and pops from both ends.
dq=c.deque(data)
dq.appendleft(1)
print(dq)
```

```python
# delete one element onlt from left
j.popleft()
print(j)
```

```python
j.popright()
print(j)
```
