````python
# 29. Write a program to print a **pyramid pattern**

# ```
#    *
#   ***
#  *****
# *******
# ```

# ---

for i in range(5,0,-1):
    for j in range(i):
        print(" ", end = "")
    for j in range(5-i):
        print("*", end = "")
    for j in range(4-i):
        print("*", end = "")
    print()
    
````

````python
# 30. Write a program to **flatten a nested list**.

# Example

# Input

# ```
# [1, [2,3], [4,[5,6]]]
# ```


# Output

# ```
# [1,2,3,4,5,6]
# ```

# ---

l = [1, [2,3], [4,[5,6]]]
s = []
for i in range(3):
    for j in range(2):
        for k in range(1):
            s.append(l[i,j,k])
            
````
