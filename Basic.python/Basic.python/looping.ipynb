```python

# butterfly 1.
n=6
for i in range(1,n):
    for j in range(i):
        print("*",end="")
    for k in range(n-i):
        print (" ",end=" ")
    for l in range(i):
        print ("*",end="")
    print()
for i in range(n,0,-1):
    for j in range(i):
        print("*",end="")
    for k in range(n-i):
        print (" ",end=" ")
    for l in range(i):
        print ("*",end="")
    print()
```
