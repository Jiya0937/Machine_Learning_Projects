```python
#Task 1: Calculate Net Price with Discount
#Description: You are given a product with a price of Rs. 500 and a discount of 10%. Calculate the net price using only arithmetic operators.

price=500
discount = 10/100
discount_money = price*discount
after_discount = price-discount_money
print("actual price of product     : ".title(),price)
print("Discount                    : ".title(),discount)
print("discount in rupees          : ".title(),discount_money)
print("actual price of the product : ".title(),after_discount)
```

```python
#Task 2: Calculate GST Amount
#Description: The price of an item is Rs. 850. You need to calculate the GST amount at 18%.

price=850
gst = 18/100
gst_money= price*gst
after_add_gst = price+gst_money
print("actual price of product     : ".title(),price)
print("gst                         : ".title(),gst)
print("gst in rupees               : ".title(),gst_money)
print("actual price of the product : ".title(),after_add_gst)
```

```python
#Task 8: Final Salary Calculation
#Description: Base salary is Rs. 30,000. Add bonus Rs. 2,000 and deduct PF Rs. 1,200 using arithmetic and assignment operators.

salary=30000
bonus=2000
PF=1200
print("salary         :".title(),salary)
print("bonus          :".title(),bonus)
print("pf             :".title(),PF)
salary+=bonus
salary-=PF
print ("actual salary  :".title(),salary)
```
