# basic string formation 


```python


#concatenation A+B
name="jiya"
lastname="deshmukh"
collage="acropolis"
city="indore"
print(collage+ " " +city)
print (name+lastname)
print("mumma " + "papa")

```

```python
#Repetition A*B
digit=3
print (name*3)
n="acropolis "
print(n*5)
```

```python
#indexing [0-1-2-3-4----n-1]
print (name[2],name[3],lastname[2])
print(n[0],n[1],n[2],n[3])
```

```python
#slicing [1:4]
print(lastname[3:5])
print(n[0:4])
print(n[4:9])
print(n[0:4]+n[4:9])
```

```python
#upper()
print (name.upper())
print(n.upper())
```

```python
#LOWER()

girl="sandhya"
boy="ADITYA"
print (girl.upper())
print (boy.lower())
print(n.lower())
```

```python
(girl),(boy)
```

```python
#title()

para="my collage name is acropolis institute of technology and research"
print(name.title(),lastname.title(), para.title())
print(n.title())
```

```python
#capitalize()
print (para.capitalize())
word=("tommorow is not a holiday and today is also a working day")
print(word.upper())
print(word.lower())
print(word.title())
print(word.capitalize())
```

```python
#strip()
print (para.strip())
print(n.strip())
```

```python
#replace(old,new)
year= "i am in 1st year"
print(year.replace('1st','2nd'))
school="career academy"
print(school.replace("academy","acropolis"))
```

```python
#find index()
name="jiya"
lastname="deshmukh"
girl="sandhya"
print(name.find('iy'),lastname.find('d'),girl.find('a'))
```

```python
# cound()  cound any letter in string
print (girl.count('a'),para.count("a"))
print(school.count("a"))
print(word.count("s"))
```

```python
#starswith()- find the world or char in present in string or not
print(name.startswith('ya'),girl.startswith("sandh"))
print(f"jiya deshmukh")
print(word.startswith("tommorow"))
print(school.startswith("academy"))
```

```python
#endswith()
print(para.endswith("collage"),para.endswith("research"))
print(school.endswith("academy"))
```

```python
#split()
letters="a,b,n,m,c,f"
print (para.split("--"), letters.split(","))

```

```python
#join()
letter=['a', 'b', 'n', 'm', 'c', 'f']
print('='.join(letter))
```

```python
#isalpha()
print("jiya".isalpha(),girl.isalpha(),"1234".isalpha())
```

```python
#isdigit
print(f"jiya190107".isdigit())
print(f"9981".isdigit())
```

```python
#isnumerical
id="0827AL241063"
print(id.isnumeric(),"2334".isnumeric(),"234.56".isnumeric())
```

```python
names="jiyadeshmukh"
print(f"my name is".startswith('m'))
print("my name is ".startswith('my'))
```
