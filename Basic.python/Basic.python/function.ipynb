```python
#recursion
def show(n):
    if n==0: #base case(for stoping program)
        return
    print(n)
    show(n-1)#recursion function                                     
    
show(9)
```

```python
def factorial(n):
    if n==0:
        return
    else:
        return factorial(n- 1)*n
print(factorial(7))
```

```python
#by function
def sum(n):
    s=0
    for i in range (1,n+1):
        s+=i
    print(s)   
sum(6)
```

```python
# by recursion
def sum(n):
    if n==0:
        return
    return sum(n-1)+n
    
add=sum(5)
print(add)
    
```
