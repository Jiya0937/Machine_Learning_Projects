# 3. Comparison Operators

```python
# == condition
x=5
y=5
if(x==y):
    print(" equal")
else:
    print("not equal")
```

```python
# != condition
x=5
y=10
if(x!=y):
    print("not equal")
else:
    print("equal")
```

```python
# > condition
x=5
y=10
if(x>y):
    print("yes")
else:
    print("no")
```

```python
# < condition
x=5
y=10
if(x<y):
    print("yes")
else:
    print("no")
```

```python
#  condition
x=5
y=10
if(x<=y):
    print("yes")
else:
    print("no")
```

```python
# >= condition
x=5
y=10
if(x>=y):
    print("yes")
else:
    print("no")
```
