```python
x=int(input("enter the no."))
m=x//500

p=x%500
r=p//200

y=p%200
n=y//100

j=y%100
a=j//50

f=j%50
b=f//20

t=f%20
h=t//10

v=t%10
w=v//5

z=v%5
q=z//2

c=z%2
d=c//1

print(f"500 : {m}")
print(f"200 : {r}")
print(f"100 : {n}")
print(f"50  : {a}")
print(f"20  : {b}")
print(f"10  : {h}")
print(f"5   : {w}")
print(f"2   : {q}")
print(f"1   : {d}")
```

```python
#nested loop


for i in range (0,2):
    for j in range (0,5):
        print(j,end=" ")
    print()
    
        
```

```python
for i in range (0,4):
    for j in range (0,4):
        print("*",end=" ")
    print(end="\n")
    
        
```

```python
n=1

for i in range (0,5):
    for j in range (0,i):
        print(n,end=" ")
        n=n+1
    print(end="\n")
    
        
```
