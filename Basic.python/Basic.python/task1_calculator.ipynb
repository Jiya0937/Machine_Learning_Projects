```python
def num(a,b):
    op=input(f"Enter operator")
    if op=="+":
        sum=a+b
        return sum
    elif op=="-":
        sub=a-b
        return sub
    elif op=="*":
        multi=a*b
        return multi
    elif op=="/":
        div=a/b
        return div
    elif op=="%":
        mod=a%b
        return mod
   
     

num(input(f"enter two variable"))

    
```
