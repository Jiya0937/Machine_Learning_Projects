```python
"""Create a program for a school where both Teacher and Student share some common details like name, 
but each has extra unique details (Teacher → subject, Student → grade). 
Use inheritance to avoid repeating the code."""
class student:
    def info(self,name,roll_no,grade):
        self.name=name
        self.roll_no=roll_no
        self.grade=grade
    def show_info(self):
        print("Name     : ",self.name)
        print("Roll NO. : ",self.roll_no)
        print("Grade    : ",self.grade)
s1=student()
s1.info("jiya",63,"A+")

class teacher(student):
    def info2(self,name,sub,dep):
        self.name=name
        self.sub=sub
        self.dep=dep
    def show(self):
        print("Name : ",self.name)
        print("Subject  : ",self.sub)
        print("Department  : ",self.dep)
t1=teacher()
t1.info2("shushma","physics","aiml")
print("======student info======")
s1.show_info()
print("======teacher info======")
t1.show()

```

```python
"""You need to create a system for an online store where Electronics and Clothing are product categories. 
Every product has a name and price, but Electronics have a warranty, and Clothing has a size."""
class products:
    def product(self,cloth,device):
        self.cloth=cloth
        self.device=device
    def cloth_product(self):
        print("Cloths : ",self.cloth)
    def device_product(self):
        print("Devices   : ",self.device)
p1=products()
p1.product("shirt","laptop")


class attributes(products):
    def attribute(self,warranty,cloth_prize,device_prize,size):
        self.warranty=warranty
        self.cloth_prize=cloth_prize
        self.device_prize=device_prize
        self.size=size
    def detail_cloth(self):
        print("prize  : " ,self.cloth_prize)
        print("size   : ",self.size)
        print("-------------------")
    def detail_device(self):  
        print("prize     : ", self.device_prize)
        print("warrranty : ",self.warranty)
a1=attributes()
a1.attribute("2 year",900,100000,"XL")
p1.cloth_product()
a1.detail_cloth()
p1.device_product()
a1.detail_device()
```

```python
"""Design a transport management program where both Bus and Train share vehicle_number and capacity. 
Buses have a route_number and trains have a track_number."""
class transport:
    def vehicles(self,v1,v2):
        self.v1=v1
        self.v2=v2
    def sow(self):
        print("vehicle :",self.v1 )
    def sow2(self):
        print("vehicle :", self.v2 )
t1=transport()
t1.vehicles("Bus","Train")
class detail_vehicles:
    def detail(self,capacity_v1,capacity_v2,route_no,track_no):
        self.capacity_v1=capacity_v1
        self.capacity_v2=capacity_v2
        self.route_no=route_no
        self.track_no=track_no
    def resource1(self):
        print("capacity : ",self.capacity_v1)
        print("Route no.: ",self.route_no)
        print("---------------")
    def resource2(self):
        print("capacity : ",self.capacity_v2)
        print("Track no.: ",self.track_no)
d1=detail_vehicles()
d1.detail(60,300,689,789)
t1.sow()
d1.resource1()
t1.sow2()
d1.resource2()
```

```python
#A freelancer platform project needs a Freelancer class that combines information from both Person (name) and Work (job role).
class name:
    def worker_name(self,name):
        self.name=name
    def show(self):
        print("Name      : ",self.name)
n1=name()
n1.worker_name("jiya")

class work:
    def rollplay(self,work):
        self.work = work
    def show2(self):
        print("Roll play :",self.work)
w1=work()
w1.rollplay(["Data Handeling","Data Analytics","Python Developer"])

class freelancer(name,work):
    def info(self):
        print("this is freelancer platform project".title())
        print("-----------=============------------")
f1=freelancer()
f1.info()
n1.show()
w1.show2()
```

```python
#In a school management system, Person class should store general details, Student should inherit Person to add grade details,
#and Graduate should inherit Student to add degree information.
class person:
    def s_info(self,name,rollno):
        self.name=name
        self.rollno=rollno
        print("Name  : ",self.name)
        print("rollno:",self.rollno)
p1=person()

class student(person):
    def grade(self,grade):
        self.grade=grade
        print("Grade : ",self.grade)
s1=student()

class graduate(student):
    def degree(self,degree,pass_year):
        self.degree=degree
        self.pass_year=pass_year
        print("Degree       : ",self.degree)
        print("Passing year : ",self.pass_year)
g1=graduate()
p1.s_info("mitansh",24)
s1.grade("A+")
g1.degree("AIML","2024-2028")
```
