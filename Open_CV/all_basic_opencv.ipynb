```python
import cv2
image=cv2.imread("C:\\image1st\\New folder\\circle.png")
if image is None:
    print("not find")
else:
    grey=cv2.cvtColor(image , cv2.COLOR_BGR2GRAY)
    # This helps separate the shape from the background.
    _,thresh=cv2.threshold(grey,120,255,cv2.THRESH_BINARY)

    #find contours
    contours,hierarchy=cv2.findContours(thresh,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)

    cv2.drawContours(image,contours,-1,(0,200,78),3)

    for contour in contours:
        approx=cv2.approxPolyDP(contour,0.01*cv2.arcLength(contour,True),True)
        corners=len(approx)
        if corners==3:
            shape_name ="Triangle"
        elif corners==4:
            shape_name= "Rectangle"
        elif corners>5:
            shape_name="circle"
        else:
            shape_name="unknown"
        cv2.drawContours(image,approx,0,(0,255,0),2)
        x=approx.ravel()[0]
        y=approx.ravel()[1]-10
        cv2.putText(image,shape_name,(x,y),cv2.FONT_HERSHEY_COMPLEX,0.6,(255,99,78),2)
    cv2.imshow("contours",image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()


```

```python
#types of bitwise operator
"""cv2.bitwise_and(img1,img2)
cv2.bitwise_or(img1,img2)
cv2.bitwise_not(img1,img2)"""
#result = cv2.bitwise_and(src1, src2, mask=None)

import cv2
import numpy as np

img1=np.zeros((300,300),dtype=np.uint8)
img2=np.zeros((300,300),dtype=np.uint8)

cv2.circle(img1,(150,150),100,255,-1)
cv2.rectangle(img2,(100,100),(250,250),255,-1)

bitwise_and=cv2.bitwise_and(img1,img2)
bitwise_or = cv2.bitwise_or(img1, img2)
bitwise_not = cv2.bitwise_not(img1)

cv2.imshow("circle",img1)
cv2.imshow("rectangle",img2)
cv2.imshow("AND",bitwise_and)
cv2.imshow("OR",bitwise_or)
cv2.imshow("NOT",bitwise_not)
cv2.waitKey(0)
cv2.destroyAllWindows()



```
