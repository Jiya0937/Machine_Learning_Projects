```python
import cv2
import numpy as np
```

```python
# image=cv2.imread()
image=cv2.imread("C:\\image1st\\New folder\\block1_sharper.jpg")
hsv=cv2.cvtColor(image,cv2.COLOR_BGR2HSV)

lower_brown1 = np.array([10,100,20])
upper_brown1 = np.array([20,255,200])

lower_brown2 = np.array([20,100,20])
upper_brown2 = np.array([30,255,200])

mask1=cv2.inRange(hsv,lower_brown1,upper_brown1)
mask2=cv2.inRange(hsv,lower_brown2,upper_brown2)
mask=mask1+mask2

cv2.imshow("Brown image",mask)
cv2.waitKey(0)

```
