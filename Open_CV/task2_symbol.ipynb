```python
#tadk 2
import cv2
image_path=input("give me a image or image path : ")
image=cv2.imread(image_path)
if image is None:
    print("image is not found")
else:
    print("1.circle\n2.rectangle\n3.square\n4.changing colour(grey)")
    user=int(input("what you want to choose : "))
    if user==1:
        #center
        x=int(input("enter x axis for center"))
        y=int(input("enter y axis for center "))
        radius=int(input("enter radius"))
        b=int(input("enter te blue color"))
        g=int(input("enter te green color"))
        r=int(input("enter te red color"))
        thickness=int(input("enter thickness"))
        cv2.circle(image,(x,y),radius,(b,g,r),thickness)
        cv2.imshow("circular image",image)
        cv2.waitKey(0)
        cv2.destroyAllWindow()
    elif user==2:
        p1_x=int(input("enter x axis for first point"))
        p1_y=int(input("enter y axis for first point"))
        p2_x=int(input("enter x axis for second point"))
        p2_y=int(input("enter y axis for second point"))
        b=int(input("enter te blue color"))
        g=int(input("enter te green color"))
        r=int(input("enter te red color"))
        thickness=int(input("enter thickness"))
        cv2.rectangle(image,(p1_x,p1_y),(p2_x,p2_y),(b,g,r),thickness)
        cv2.imshow("circular image",image)
        cv2.waitKey(0)
        cv2.destroyAllWindow()
    elif user==3:
        p=int(input("enter starting point"))
        p1=p*2
        b=int(input("enter te blue color"))
        g=int(input("enter te green color"))
        r=int(input("enter te red color"))
        thickness=int(input("enter thickness"))
        cv2.rectangle(image,(p,p),(p1,p),(b,g,r),thickness)
        cv2.imshow("circular image",image)
        cv2.waitKey(0)
        cv2.destroyAllWindow()
    elif user==4:
        #CONVERT INTO GREY SCALE
        gray_image=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)

        #SHOE BOTH NORMAL AND GREY IMAGE
        cv2.imshow("original image",image)
        cv2.imshow("grey image",gray_image)
        cv2.waitKey(0)
        cv2.destroyAllWindow()
        
        
        
        
        
```
