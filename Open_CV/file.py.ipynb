```python
import cv2
import numpy as np

# --- Function to find X and Y centroid ---
def get_centroid(contour):
    M = cv2.moments(contour)
    if M["m00"] != 0:
        cx = int(M["m10"] / M["m00"])
        cy = int(M["m01"] / M["m00"])
        return (cx, cy)
    else:
        return None

# --- Load image ---
img = cv2.imread("block2_image.jpg")
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# --- Threshold to get binary mask ---
_, thresh = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)

# --- Find contours ---
contours, _ = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# --- Find centroids of all contours ---
centroids = []
for c in contours:
    centroid = get_centroid(c)
    if centroid:
        centroids.append((centroid[0], centroid[1], c))  # (x, y, contour)

# --- Sort contours top-to-bottom (by Y) then left-to-right (by X) ---
centroids.sort(key=lambda x: (x[1] // 100, x[0]))  # 100 = row height approx adjustment

# --- Draw sorted contours + assign IDs ---
row_count = 2
col_count = 3
font = cv2.FONT_HERSHEY_SIMPLEX

for idx, (x, y, c) in enumerate(centroids):
    row = idx // col_count + 1
    col = idx % col_count
    plant_id = f"p{row}{chr(65 + col)}"
    cv2.drawContours(img, [c], -1, (0, 255, 0), 3)
    cv2.putText(img, plant_id, (x - 30, y - 20), font, 1.5, (0, 0, 255), 3)

# --- Show output ---
cv2.imshow("Plant IDs (2x3 Grid)", img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
