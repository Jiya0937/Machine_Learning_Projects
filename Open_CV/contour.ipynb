```python
import cv2
import numpy as np

image = cv2.imread("block1_image.jpg")



if image is None:
    print("No image")
    exit()

gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
ret, thresh = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)

contours, _ = cv2.findContours(thresh, cv2.RETR_LIST, cv2.CHAIN_APPROX_NONE)

# Filter contours based on area
filtered_contours = [c for c in contours if cv2.contourArea(c) > 500]

print("Number of large contours = ", len(filtered_contours))
remaining = len(contours) - len(filtered_contours)
print("Number of small contours filtered out = ", remaining)

# Draw contours
cv2.drawContours(image, filtered_contours, -1, (0, 255, 0), 1)

# Assign IDs to filtered contours
plant_names = ['p1A', 'p2B', 'p3C', 'p4D', 'p5E', 'p6F']

for idx, cnt in enumerate(filtered_contours):
    if idx < len(plant_names):
        # Find centroid to place the text
        M = cv2.moments(cnt)
        if M["m00"] != 0:
            cx = int(M["m10"] / M["m00"])
            cy = int(M["m01"] / M["m00"])
        else:
            cx, cy = 0, 0
        
        cv2.putText(image, plant_names[idx], (cx-20, cy), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 255), 1)

cv2.imshow("Contours with IDs", image)
cv2.imshow("Gray", gray)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
image=cv2.imread("C:\\image1st\\New folder\\shapes.png")
gray=cv2.cvtColor(image,cv2.COLOR_BGE2GRAY)

contours,hierarchy = cv2.findContours(gray,cv2.RETR_CCOMT,cv2.CHAIN_APPROX_NONE)

sorted_contors = sorted(contours, key=cv2.contourArea,reverse=True)

for i, cont in enumerate(sorted_contours[:3],1):
    cv2.drawContour(image,cont,-1,(0,255,0),3)

    cv2.putText(image,str(i),(cont[0,0,0],cont[0,0,1]-10),cv2.FONT_HERSHEY_SIMPLEX,1.4,(34,34,34),3)

cv2.imshow("contour image",image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
