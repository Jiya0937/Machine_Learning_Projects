```python
import cv2
face_cascade=cv2.CascadeClassifier("C:\\Users\\jiyad\\PycharmProjects\\PythonProject\\finding and dawing contours in opencv\\haarcascade_frontalface_alt.xml")
smile_cascade=cv2.CascadeClassifier("C:\\Users\\jiyad\\PycharmProjects\\PythonProject\\finding and dawing contours in opencv\\haarcascade_smile.xml")
eye_cascade=cv2.CascadeClassifier("C:\\Users\\jiyad\\PycharmProjects\\PythonProject\\finding and dawing contours in opencv\\haarcascade_eye.xml")

cap=cv2.VideoCapture(0)
while True:
    ret, frame=cap.read()
    if not ret:
        break
    gray=cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)

    faces=face_cascade.detectMultiScale(gray,1.1,5)

    for (x,y,w,h) in faces:
        cv2.rectangle(frame,(x,y),(x+w,y+h),(252,77,99),2)

        roi_gray = gray[y:y+h,x:x+w]

        roi_color = frame[y:y+h,x:x+w]

        eyes = eye_cascade.detectMultiScale(roi_gray, 1.1, 10)
        if len(eyes) > 0:
            cv2.putText(frame, "show eyes", (x, y - 10), cv2.FONT_HERSHEY_COMPLEX, 1.1,(255, 255, 0), 3)
        smile = smile_cascade.detectMultiScale(roi_gray, 1.1, 10)
        if len(smile) > 0:
            cv2.putText(frame, "smiling ", (x, y - 30), cv2.FONT_HERSHEY_COMPLEX, 1.1,(255, 0, 0), 3)
        cv2.imshow("smart face detection",frame)
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break

cap.release()
cv2.destroyAllWindows()
"""
x=100
y=150
w=80
h=80
"""
```
