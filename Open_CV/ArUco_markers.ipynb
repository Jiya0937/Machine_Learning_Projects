```python
import cv2
import cv2.aruco as aruco

VideoCap = False
cap = cv2.VideoCapture(0)

def findAruco(image, marker_size=6, total_markers=250, draw=True):
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Choose dictionary
    key = getattr(aruco, f'DICT_{marker_size}X{marker_size}_{total_markers}', None)
    if key is None:
        print("INVALID ARUCO DICTIONARY")
        return None, None

    arucoDict = aruco.getPredefinedDictionary(key)
    aruco_param = aruco.DetectorParameters()

    # Detector
    detector = aruco.ArucoDetector(arucoDict, aruco_param)

    # Detect markers
    bbox, ids, rejected = detector.detectMarkers(gray)

    if ids is not None and len(ids) > 0:
        print("Detected IDs:", ids.flatten())
        if draw:
            aruco.drawDetectedMarkers(image, bbox, ids)
        return bbox, ids
    else:
        print("No ArUco markers detected")
        return None, None


while True:
    if VideoCap:
        ret, image = cap.read()
        if not ret:
            break
    else:
        image = cv2.imread("C:\\image1st\\New folder\\a1.png")
        if image is None:
            print("Image not found")
            break
        print("Image is found")
        image = cv2.resize(image, (0, 0), fx=0.5, fy=0.5)

        findAruco(image)

        cv2.imshow("aruco image", image)
        cv2.waitKey(0)
        cv2.destroyAllWindows()
        break

cap.release()
cv2.destroyAllWindows()
```
