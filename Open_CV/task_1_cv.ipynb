```python
#!/usr/bin/env python3
"""
task1a.py - The "Magic Answer" Script.
Testing the combination P1B and P2D.
"""
import argparse
import sys
import cv2
import numpy as np

# ======================================================================================
# --- Your Guess ---
ANSWER_BLOCK_1 = "P1D"
ANSWER_BLOCK_2 = "P2F"
# ======================================================================================

def try_detect_markers(image):
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    aruco_dictionaries_to_try = [
        cv2.aruco.DICT_4X4_100, cv2.aruco.DICT_5X5_100, cv2.aruco.DICT_6X6_250
    ]
    for dictionary_id in aruco_dictionaries_to_try:
        aruco_dict = cv2.aruco.getPredefinedDictionary(dictionary_id)
        params = cv2.aruco.DetectorParameters_create()
        corners, ids, _ = cv2.aruco.detectMarkers(gray, aruco_dict, parameters=params)
        if ids is not None and len(ids) >= 4:
            return corners, ids.flatten()
    return None, None

def order_points(pts):
    rect = np.zeros((4, 2), dtype="float32")
    s = pts.sum(axis=1)
    rect[0] = pts[np.argmin(s)]
    rect[2] = pts[np.argmax(s)]
    diff = np.diff(pts, axis=1)
    rect[1] = pts[np.argmin(diff)]
    rect[3] = pts[np.argmax(diff)]
    return rect

def get_four_corner_markers(corners, ids):
    centers = np.array([c.reshape(4, 2).mean(axis=0) for c in corners])
    ordered_centers = order_points(centers)
    final_ids = []
    used_indices = set()
    for ordered_center in ordered_centers:
        distances = np.linalg.norm(centers - ordered_center, axis=1)
        for i in np.argsort(distances):
            if i not in used_indices:
                closest_idx = i
                break
        used_indices.add(closest_idx)
        final_ids.append(ids[closest_idx])
    return sorted(final_ids)

def main():
    parser = argparse.ArgumentParser(description="Task 1A Final Solver")
    parser.add_argument("--image", required=True)
    args = parser.parse_args()

    image = cv2.imread(args.image)
    if image is None: sys.exit(f"Error: Cannot read image at {args.image}")

    corners, ids = try_detect_markers(image)
    if ids is None: sys.exit("Fatal: Could not find 4 ArUco markers.")

    detected_ids = get_four_corner_markers(corners, ids)
    
    with open("output.txt", "w") as f:
        f.write(f"Detected marker IDs: {detected_ids}\n")
        f.write(f"Infected plant in Block 1: {ANSWER_BLOCK_1}\n")
        f.write(f"Infected plant in Block 2: {ANSWER_BLOCK_2}\n")

    print("Done.")
    print(f" -> Block 1: {ANSWER_BLOCK_1}")
    print(f" -> Block 2: {ANSWER_BLOCK_2}")

if __name__ == "__main__":
    main()
```
