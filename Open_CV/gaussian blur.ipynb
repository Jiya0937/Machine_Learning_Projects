```python
import cv2
```

```python
#blurred image : reduce noise and detail in an image.
image=cv2.imread("C:\\Users\\jiyad\\Downloads\\eye.png")
if image is None:
    print("image is not loaded")
else:
    blurred=cv2.GaussianBlur(image,(9,7),1)
    cv2.imshow("original image",image)
    cv2.imshow("blurred image",blurred)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```
