```python
import cv2
import os

# Folder create karna
distances = ["10ft", "20ft", "30ft", "40ft"]
base_path = "person_dataset"

for d in distances:
    os.makedirs(os.path.join(base_path, d), exist_ok=True)

# Webcam start karo
cap = cv2.VideoCapture(0)  # Agar external camera hai toh index change karo

print("Press 's' to save image, 'q' to quit")

distance_index = 0
img_count = 0

while True:
    ret, frame = cap.read()
    if not ret:
        break

    # Show live camera feed
    cv2.imshow("Camera Feed", frame)

    key = cv2.waitKey(1) & 0xFF

    # Agar 's' dabaya toh image save hogi
    if key == ord('s'):
        folder = os.path.join(base_path, distances[distance_index])
        img_name = f"img_{img_count}.jpg"
        cv2.imwrite(os.path.join(folder, img_name), frame)
        print(f"Saved {img_name} in {distances[distance_index]}")
        img_count += 1

    # Agar 'n' dabaya toh next distance par chale jao
    elif key == ord('n'):
        distance_index = (distance_index + 1) % len(distances)
        img_count = 0
        print(f"Now capturing for distance: {distances[distance_index]}")

    # Quit
    elif key == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
