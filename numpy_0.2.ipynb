```python
"""7. NumPy Array Iterating
Iterating means going through elements one by one.
As we deal with multi-dimensional arrays in numpy, we can do this using basic for loop of python.
If we iterate on a 1-D array it will go through each element one by one."""

arr=np.array([1,2,3])
for x in arr:
    print(x)
```

```python
#Iterating 2-d array
arr=np.array([[12,3,4],[3,4,5]])
for x in arr:
    print(x)
    
```

```python
#Splitting NumPy Arrays
l=np.array([2,3,4,5,6,888,9,3])
s=np.array_split(l,3)
print(s)
```

```python
##Searching array(where)
arr=np.array([4,4,5,6,67,8])
print(np.where(arr==67))
print(np.where(arr==0))
```

```python
#sorting arrays
arr=np.array([3,4,5,0,7])
print(np.sort(arr))
```
