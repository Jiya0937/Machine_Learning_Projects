# Encode Categorical Values

```python
data = {
    'Name': ['Aditi', 'Rohan', 'Priya', 'Arjun', 'Sneha', 'Karan', 'Neha', 'Raj'],
    'Gender': ['Female', 'Male', None, 'Male', 'Female', 'Male', 'Female', None],
    'Department': ['CSE', 'AIML', 'IT', None, 'ECE', 'IT', 'AIML', 'CSE'],
    'Age': [19, None, 18, 21, 20, None, 22, 20],
    'CGPA': [8.5, 7.8, None, 7.2, 8.8, 8.1, 9.0, None],
    'City': ['Indore', 'Bhopal', 'Ujjain', 'Indore', None, 'Pune', 'Bhopal', 'Indore']
}
```

```python
dataset = pd.DataFrame(data)
dataset
```

```python
dataset.isnull().sum()
```

```python
dataset.isnull().sum().sum()
```

```python
dataset.select_dtypes(include = 'float64')
```

```python
print(dataset['Age'].mean())
print(dataset['CGPA'].mean())
```

```python
# fill numerical nulll value by call sklearn library
```

```python
from sklearn.impute import SimpleImputer
```

```python
si = SimpleImputer(strategy = "mean")
```

```python
dataset[['Age', 'CGPA']] = si.fit_transform(dataset[['Age', 'CGPA']])
dataset
```

```python
# decrease moredecimal value after point in 2 value only
dataset['CGPA'] = dataset['CGPA'].round(2) 
```

# 3. Feature Scaling

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler , MinMaxScaler
from sklearn.model_selection import train_test_split
```

```python
data = {
    'Study Hours' : [1,2,3,4,5,6],
    'Test Scores' : [20,30,40,50,60,70]
}

df=pd.DataFrame(data)

#standardscaler
standard_scaler = StandardScaler()
standard_scaled = standard_scaler.fit_transform(df)

print("Standarf Scaled Output : ")
print(pd.DataFrame(standard_scaled, columns = ['Study Hours' , 'Test Score' ]))

#minaxScaler
minmax_scaler = MinMaxScaler()
minmax_scaled = minmax_scaler.fit_transform(df)

print("\n\n MinMax Scaled Output : ")
print(pd.DataFrame(minmax_scaled,columns = ['Study Hours','TestScore']))
```

# 4. Split Data

```python
x = df[['Study Hours']]
y = df[['Test Scores']]

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2,random_state=2)
print("\nTrainig Data")
print(x_train)

print("\nTesting Data")
print(x_test)

print("\nTraining Data")
print(y_train)

print("\nTesting Data")
print(y_test)
```
