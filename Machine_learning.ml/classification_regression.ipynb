# Logistic Regression

```python
# logistic regression

from sklearn.linear_model import LogisticRegression
X = [[4], [5], [6], [8]]
y = [0,0,1,1]

model = LogisticRegression()
model.fit(X,y)
hours = float(input("enter your housrs yor studies : "))
result = model.predict([[hours]])[0]

if result == 1:
    print(f"Based on hours {hours} , you are likely PASS")
else:
    print(f"Based on hours {hours} , you are likely FAIL")
    
```

# KNN Regression

```python
from sklearn.neighbors import KNeighborsClassifier

X = [
    [100,8],
    [120,9],
    [110,10],
    [300,4],
    [400,5],
    [50,3]
]

# 0 = banana, 1= Orange
y = [0,0,0,1,1,0]

model = KNeighborsClassifier(n_neighbors = 3)
model.fit(X,y)

weight = float(input("Enter the weight of fruit in grams"))
size = float(input("Enter the size of fruit in centimeter"))

prediction = model.predict([[weight, size]])[0]

if prediction == 0:
    print("your fruit is Banana")
else:
    print("your fruit is Apple")
```

```python
from sklearn.tree import DecisionTreeClassifier

X = [
    [7,2],  #Apple
    [8,3],  #Apple
    [9,8],  #Orange
    [10,9]  #Orange
]
y = [0,0,1,1]  # 0 = Apple , 1 = Orange

model = DecisionTreeClassifier()

model.fit(X,y)

size = float(input("Enter the size of the fruit in cm "))
shade = float(input("Enter the shade of the fruit in bgr "))

prediction = model.predict([[size,shade]])[0]

if prediction == 0:
    print("it maybe Apple ")

else:
    print("it maybe orange")
             
 
```

```python
'''
# Overfitting :- When a model learns too much from the training data — even the noise and outliers —
it performs very well on training data but poorly on new data.

Why it happens:
Model is too complex (too many parameters or layers)
Too much training or too few data points
No regularization

# Under Fitting :- When a model is too simple and fails to learn important patterns in the training data.
Why it happens:
Model is not complex enough
Too little training
Missing important features

# Good Fitting :- When the model learns the right patterns — not too much, not too little.
It performs well on both training and test data.

'''
```

```python
from sklearn.metrics import accuracy_score,precision_score, recall_score, f1_score

#True answere (what actually happened)
y_true = [1, 0, 1, 1, 0, 1, 0]

#Model's Prediction (what it guessed)
y_pred = [1, 0, 1, 0, 0, 1, 1]

# Evaluation 
print("Accuracy : ", accuracy_score(y_true, y_pred)) # model  71 to 75%  predict right 
print("Precision : ", precision_score(y_true, y_pred))
print("Recall : ",recall_score(y_true, y_pred))
print("F1 score: ", f1_score(y_true, y_pred))
```

# Confusion matrix

```python
# for visvalise true_posotive, false_positive, false_positive, false_negative
# use confusion matrix 

'''Acual true| Actual False
   __________|____________  
  | true_pos | false_pos |
  |__________|___________| 
  | false_neg| true_neg  |        
  |__________|___________|   

  A Confusion Matrix in machine learning is a
table used to evaluate the performance of a classification model.
It shows how many predictions your model got correct and how many 
it got wrong, divided into different categories.

📌 Meaning of Each Term
1️⃣ True Positive (TP)

Model predicted Positive, and it is actually Positive.
✔ Correct prediction.

2️⃣ True Negative (TN)

Model predicted Negative, and it is actually Negative.
✔ Correct prediction.

3️⃣ False Positive (FP)

Model predicted Positive, but it is actually Negative.
❌ Wrong prediction.
(Also called Type-I Error)

4️⃣ False Negative (FN)

Model predicted Negative, but it is actually Positive.
❌ Wrong prediction.
(Also called Type-II Error)
'''
```

```python

from sklearn.metrics import confusion_matrix

y_true = [1,0,1,1,0,1,0,1,0,0]
y_pred = [1,0,0,1,0,1,0,0,1,0]

cm = confusion_matrix(y_true, y_pred)

print("confusion matrix")
print(cm)
```

```python
# Mean Absolute Error (MAE)
# Mean Square Error (MSE)
# Root Mean Square Error (RMSE)
```

```python
# Absolute Error = | Actual Value – Predicted Value |
# It always gives a positive error because of the absolute ( | | ) sign.
'''
📌 Example

Suppose:

Actual value = 50
Predicted value = 45

Then,
Absolute Error = |50 − 45| = 5

----When Absolute Error is used?----

Regression problems
Forecasting
Evaluating models like Linear Regression, Decision Tree, Random Forest, etc.
'''
```

```python
# MSE measures the average of the squared differences between the actual values and the predicted values.
'''
📊 Example : 

Actual values: [3, 5, 2]
Predicted values: [4, 4, 1]

Errors =
(3−4)² = 1
(5−4)² = 1
(2−1)² = 1

MSE=(1+1+1)/3=1
----------------------------------
 *  Where MSE is used?
Linear Regression
Polynomial Regression
Gradient Descent (as a loss function)
Forecasting models
Any regression model performance measurement

'''
```

```python
# RMSE is the square root of the Mean Square Error (MSE)
'''
========What RMSE tells you?==========
It shows how much error your model makes on average, in the same units as the actual data.
 * If you predict house price in rupees → RMSE is also in rupees
 * If you predict temperature → RMSE is in °C
This makes RMSE easier to interpret than MSE. 

----------------------------------------------
📊 Example

Actual values: [3, 5, 2]
Predicted values: [4, 4, 1]

Squared errors:
1, 1, 1
MSE = (1+1+1)/3 = 1
RMSE = √1 = 1

-----------------------------------------------

--------------When RMSE is useful?-----------------------
When large errors are more harmful
Regression tasks like:
House price prediction
Weather prediction
Stock prediction
Demand forecasting
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np

# real scores
real_scores = [90,60,80,100,70]

#model prediction
model_predict = [85,70,75,95,60]

MAE = mean_absolute_error(real_scores, model_predict)
MSE = mean_squared_error(real_scores, model_predict)
RMSE = np.sqrt(MSE)

print("MAE : on average off by : ", MAE) # MAE use when you have to see small mistake unlarge dataset
print("MSE : on average off by : ", MSE) # MSE use when you have to find big mistake like outlier
print("RMSE : on average off by : ", RMSE) 

```
