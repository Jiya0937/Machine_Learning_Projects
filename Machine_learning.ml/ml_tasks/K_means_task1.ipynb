# Encode Categorical Values

```python
from sklearn.preprocessing import LabelEncoder
```

```python
df2 = df.copy()
```

df2

```python
df2.select_dtypes(include = "object").columns
```

```python
le = LabelEncoder()

for col in df2.select_dtypes(include = "object").columns:
    df2[col] = le.fit_transform(df2[col])
```

# Feature Scalling

```python
# StandardScaler 
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

```python
# standerd.scaler 
standard_scaler = StandardScaler()
standard_scaled = standard_scaler.fit_transform(df2)
df2_scale = round(pd.DataFrame(standard_scaled,columns=df.columns),2)
df2_scale
```

```python
# Minmax scaler
```

```python
minmax_scaler = MinMaxScaler()
minmax_scaled = minmax_scaler.fit_transform(df2_scale)
print("\n\nMinMax Scaled Output")
df_fix = pd.DataFrame(minmax_scaled, columns = df2_scale.columns)
```

# Split Training and Testing data

```python
# split data

from  sklearn.model_selection import train_test_split 

X = df_fix[['Hours_Studied', 'Attendance',	'Extracurricular_Activities',	'Sleep_Hours',	'Previous_Scores',	'Motivation_Level',	'Internet_Access',	'Tutoring_Sessions',	'School_Type',	'Gender']]
y = df_fix['Exam_Score']

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.3, random_state = 34)

```
