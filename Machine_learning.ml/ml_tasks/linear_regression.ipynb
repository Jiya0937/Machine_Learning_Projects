```python
from sklearn.linear_model import LinearRegression

X = [[4], [5], [6], [8]]
y = [50,60,70,90]

model = LinearRegression()

model.fit(X,y)  # used to train our model(learns the relationship between X and y)

hours = float(input("enter how many hours you studies"))

predicted_marks = model.predict([[hours]])  #(uses that learned relationship to make new predictions)

print(f"Based on your {hours} you may score around {predicted_marks}")
```
