```python
batsmans = {
    "Name" : ["Virat Kohli", "Rohit Sharma", "Hardik Pandiya", "Ravindra Jadeja", "Ms Dhoni"],
    '2015'   : [234,234, 564, 242, 242],
    '2016'   : [453, 342, 543, 234, 532],
    '2017'   : [533, 454, 523, 432, 543]
}
mans =pd.DataFrame(batsmans)
mans
```

```python
# arrange bar chart in the size by size
plt.bar(np.arange(mans.shape[0]) - 0.2, mans['2015'], width = 0.2, color = "black")
plt.bar(np.arange(mans.shape[0]), mans['2016'], width = 0.2, color = "brown")
plt.bar(np.arange(mans.shape[0]) + 0.2, mans['2017'], width = 0.2, color = "red")
plt.xticks(np.arange(mans.shape[0]), mans["Name"])
plt.xticks(rotation = "vertical")
plt.show
```

```python
plt.bar(mans["Name"] , mans['2017'], width = 0.6, color = "black")
plt.bar(mans["Name"], mans['2016'], bottom = mans['2017'], width = 0.6, color = "brown")
plt.bar(mans["Name"], mans['2015'], bottom = mans['2016'] + mans['2017'], width = 0.6, color = "red")
plt.xticks(np.arange(mans.shape[0]), mans["Name"])
plt.xticks(rotation = "vertical")
plt.legend(["2017", "2016", "2015"], loc = 'upper left')
plt.show()
```
